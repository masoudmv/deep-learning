```python
# Load the Cora dataset
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
dataset = Planetoid(root='data', name='Cora', transform=NormalizeFeatures())
data = dataset[0]
data = data.to(device)


train_mask = data.train_mask
x_train = data.x[train_mask]  
y_train = data.y[train_mask]  

# Reduce dimensions to 2D using t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
x_train_2d = tsne.fit_transform(x_train.cpu())


# Plot
plt.figure(figsize=(8, 6))
scatter = plt.scatter(x_train_2d[:, 0], x_train_2d[:, 1], c=y_train.cpu(), cmap='tab10', s=30, alpha=0.8)
plt.colorbar(scatter, label='Class')
plt.title('t-SNE Visualization of Cora Training Set')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.grid(True)
plt.show()
```

```python
class GCNConv(MessagePassing):
    def __init__(self, in_channels, out_channels, k=1):
        super(GCNConv, self).__init__(aggr='add')
        self.linear = nn.Linear(in_channels, out_channels)
        self.k = k  # Number of propagation steps (hops)



    def forward(self, x, edge_index):
        edge_index, _ = add_self_loops(edge_index, num_nodes=x.size(0))

        row, col = edge_index
        deg = degree(row, x.size(0), dtype=x.dtype)              
        deg_inv_sqrt = deg.pow(-0.5)                             
        norm = deg_inv_sqrt[row] * deg_inv_sqrt[col]             

        out = self.linear(x)

        for _ in range(self.k):
            out = self.propagate(edge_index, x=out, norm=norm)

        return out

        
    def propagate(self, edge_index, x, norm):
        row, col = edge_index
        msg = self.message(x_j=x[col], norm=norm)                                
        aggr_out = self.aggregate(msg, row, dim_size=x.size(0))                  
        return self.update(aggr_out)

    
    def message(self, x_j, norm):
        return norm.view(-1, 1) * x_j

    
    def aggregate(self, inputs, index, dim_size=None):
        return super().aggregate(inputs, index, dim_size=dim_size)

    
    def update(self, aggr_out):
        return aggr_out
```

```python
os.makedirs("tsne_frames", exist_ok=True)
```

```python
def save_tsne_frame(epoch, model, data):
    model.eval()
    with torch.no_grad():
        # Extract features after first GCN layer
        x = F.relu(model.layers[0](data.x, data.edge_index))
        train_mask = data.train_mask
        x_train = x[train_mask].cpu().numpy()
        y_train = data.y[train_mask].cpu().numpy()

        # Apply t-SNE
        tsne = TSNE(n_components=2, random_state=42, perplexity=30)
        x_2d = tsne.fit_transform(x_train)

        # Save plot
        plt.figure(figsize=(7, 6))
        scatter = plt.scatter(x_2d[:, 0], x_2d[:, 1], c=y_train, cmap='tab10', s=30, alpha=0.8)
        plt.colorbar(scatter, label='Class')
        plt.title(f't-SNE (Training Set)  Epoch {epoch}')
        plt.xlabel('dim 1')
        plt.ylabel('dim 2')
        plt.grid(True)
        plt.tight_layout()
        plt.savefig(f"tsne_frames/tsne_epoch_{epoch:03d}.png")
        plt.close()
```

```python
import matplotlib.animation as animation
from IPython.display import HTML
from PIL import Image

def show_tsne_animation():
    # Load images into frames
    frame_paths = sorted([f"tsne_frames/{f}" for f in os.listdir("tsne_frames") if f.endswith(".png")])
    frames = [Image.open(fp) for fp in frame_paths]

    # Create animation using matplotlib
    fig, ax = plt.subplots(figsize=(7, 6))
    img = plt.imshow(frames[0])
    plt.axis('off')

    def update(i):
        img.set_data(frames[i])
        return [img]

    ani = animation.FuncAnimation(fig, update, frames=len(frames), interval=500)
    plt.close(fig)
    return HTML(ani.to_jshtml())
```

```python
class StackedGCN(nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels, num_layers):
        super().__init__()
        
        self.layers = nn.ModuleList()

        
        if num_layers == 1:
            self.layers.append(GCNConv(in_channels, hidden_channels, k=1))
            self.layers.append(nn.Linear(hidden_channels, out_channels))
        
        else:
            self.layers.append(GCNConv(in_channels, hidden_channels, k=1))
            self.layers.append(nn.Linear(hidden_channels, hidden_channels))
            
            for _ in range(num_layers - 2):
                self.layers.append(GCNConv(hidden_channels, hidden_channels, k=1))
                self.layers.append(nn.Linear(hidden_channels, hidden_channels))
                
            self.layers.append(GCNConv(hidden_channels, hidden_channels, k=1))
            self.layers.append(nn.Linear(hidden_channels, out_channels))
        
        
    def forward(self, x, edge_index):
        for i, layer in enumerate(self.layers):
            if isinstance(layer, GCNConv):
                x = layer(x, edge_index)
            else:  
                x = layer(x)
            if i < len(self.layers) - 1:
                x = F.relu(x)
        return x
```

```python
class SingleKHopGCN(nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels, k):
        super().__init__()
        self.conv = GCNConv(in_channels, hidden_channels, k)
        self.linear = nn.Linear(hidden_channels, out_channels)
        self.layers = [self.conv, self.linear]

    def forward(self, x, edge_index):
        x = self.conv(x, edge_index)
        x = self.linear(x)
        return x
```

```python
def train(model, data , epoch_num):
    model.train()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
 
    for epoch in range(1, epoch_num):

        optimizer.zero_grad()
        out = model(data.x, data.edge_index)
        loss = nn.CrossEntropyLoss()(out[data.train_mask], data.y[data.train_mask])
        loss.backward()
        optimizer.step()
        
        if epoch % 20 == 0:
            save_tsne_frame(epoch, model, data)

def test(model, data):
    model.eval()
    out = model(data.x, data.edge_index)
    pred = out.argmax(dim=1)
    correct = pred[data.test_mask] == data.y[data.test_mask]
    return int(correct.sum()) / int(data.test_mask.sum())


K = 2  # Number of layers / hops
in_channels = dataset.num_node_features
hidden_channels = 32
out_channels = dataset.num_classes

model1 = StackedGCN(in_channels, hidden_channels, out_channels, num_layers=K).to(device)


train(model1, data, 500)
acc1 = test(model1, data)

print(f"Stacked GCN ({K} layers, k=1 each): Test Accuracy = {acc1:.4f}")
show_tsne_animation()
```

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt

Ks = range(1, 6)
accs_model1 = []
accs_model2 = []

in_channels = dataset.num_node_features
hidden_channels = 32
out_channels = dataset.num_classes

for K in Ks:

  model1 = StackedGCN(in_channels, hidden_channels, out_channels, K).to(device)
  train(model1, data, 500)
  accs_model1.append(test(model1, data))

  model2 = SingleKHopGCN(in_channels, hidden_channels, out_channels, K).to(device)
  train(model2, data, 500)
  accs_model2.append(test(model2, data))

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(Ks, accs_model1, marker='o', label='Stacked GCN (K layers, 1-hop)')
plt.plot(Ks, accs_model2, marker='s', label='Single GCN (1 layer, k-hop)')
plt.xlabel("K (number of layers / hops)")
plt.ylabel("Test Accuracy")
plt.title("Comparison of Stacked vs k-Hop GCNs")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
